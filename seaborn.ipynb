- line plot


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,6,7]
y_1=sns.load_dataset("titanic")
x_1=pd.DataFrame({"var":var,"var_1":var_1})
sns.lineplot(x="var",y="var_1",data=x_1)
plt.show()
```

```python
plt.plot(var,var_1)
plt.show()
```

```python
y_1=sns.load_dataset("titanic").head(15)
y_1
```

```python
sns.lineplot(x="age",y="fare",data=y_1,hue="sex",style="sex",size=50,palette="Accent",markers=["o",">"],dashes=False,legend="full")
plt.grid()
plt.title("python")
plt.show() 


```

- histogram plot


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("titanic").head(20)
var
```

```python
sns.displot(var["sex"],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True,color="g",log_scale=True)
plt.show()
```

- bar plot

```python
sns.barplot(x="survived",y="sex",data=var,hue="sex",color="b",palette="Accent",saturation=10,errcolor="g")
plt.show()
```

- scatter plot

```python
sns.scatterplot(x="age",y="embark_town",data=var,hue="sex",palette="gist_ncar",style="sex",size="sex",sizes=(50,60),alpha=1)
```
